```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
track = pd.read_csv('tracks.csv')
```

```python
track.head()
```

## Checking for null value

```python
pd.isnull(track).sum()
```

```python
track.info()
```

```python
track.dropna(axis =0, inplace = True)
```

```python
track.isnull().sum()
```

## 10 Least popular tracks

```python
sorted_df= track.sort_values('popularity', ascending= True).head(10)
```

```python
sorted_df
```

## Most popular artist

```python
track.groupby(by = 'artists', ).max('popularity').sort_values('popularity', ascending= False).head(1)
```

## Top 10 songs

```python
sorted_pop= track.query( 'popularity > 90', inplace = False).sort_values('popularity', ascending = False)
```

```python
sorted_pop.head(10)
```

```python
track.set_index('release_date', inplace =True)
```

```python
track.index = pd.to_datetime(track.index)
```

### Converting duration in Milliseconds to seconds

```python
track['duration'] = track['duration_ms'].apply(lambda x: round(x/1000))
```

```python
track.drop('duration_ms', axis = 1, inplace = True)
```

```python
track['duration'].head()
```

```python
track
```

## Finding correlation

```python
corr_df = track.drop(['key', 'mode', 'explicit'], axis = 1).corr(method = 'pearson')
```

```python
corr_df
```

```python
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_df, annot = True, vmin = -1, vmax = 1, cmap = 'coolwarm' )
heatmap.set_title('Correlation heat map between variables')
```

## Analysing loundness vs energy correlation 

```python
sample_df = track.sample(int(0.004*len(track)))
sample_df.info()
```

```python
plt.figure(figsize = (14,6))
sns.regplot(data = sample_df, x = 'loudness', y = 'energy', marker = 'o').set_title('loudness vs energy correlation')
```

```python
plt.figure(figsize = (14,6))
sns.regplot(data = sample_df, x = 'acousticness', y = 'popularity', color = 'b', marker = 'o').set_title('accousticness vs popularity correlation')
```

### Analysing Genre of Song and duration 

```python
df_genre = pd.read_csv('SpotifyFeatures.csv')
```

```python
df_genre.head(5)
```

```python
df_genre.info()
```

```python
df_genre.isnull().sum()
```

```python
plt.figure(figsize = (12,6))
plt.title("Duration of songs in different Genres")
sns.color_palette("rocket", as_cmap= True)
sns.barplot(y = 'genre', x = 'duration_ms', data = df_genre)
plt.xlabel('duration in ms')
plt.ylabel('genre')
```

```python
sns.set_style(style = 'darkgrid')
plt.figure(figsize = (10,5))
famous = df_genre.sort_values('popularity', ascending=False).head(10)
sns.barplot(y = 'genre', x = 'popularity', data = famous).set(title = 'top 5 genres by popularity')
```
